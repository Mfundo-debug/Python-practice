```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

def vectorized_function():
    array1 = np.random.rand(1000000)
    array2 = np.random.rand(1000000)
    return np.divide(array1, array2)

def loop_based_function():
    array1 = np.random.rand(1000000)
    array2 = np.random.rand(1000000)
    result = np.zeros_like(array1)
    for i in range(len(array1)):
        if array2[i] != 0:
            result[i] = array1[i] / array2[i]
    return result

vectorized_time = timeit.timeit(vectorized_function, number=100)
loop_time = timeit.timeit(loop_based_function, number=100)

plt.bar(["Vectorized", "Loop-based"], [vectorized_time, loop_time])
plt.ylabel("Time (s)")
plt.title("Comparison of Vectorized vs Loop-based Functions")
plt.show()
```
