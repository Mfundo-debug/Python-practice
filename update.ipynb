```python
import pandas as pd

# Load the xlsx file
df = pd.read_excel('updateddataset.xlsx', header=None)

# The data is separated by both commas and tabs. We'll need to handle this.
# First, split the DataFrame on tabs.
df = df[0].str.split('\t', expand=True)

# Then, split the columns that contain commas.
for col in df.columns:
    # Check if the column contains commas
    if df[col].str.contains(',').any():
        # Split the column
        split_col = df[col].str.split(',', expand=True)
        # Concatenate the split columns with the original DataFrame
        df = pd.concat([df.drop(col, axis=1), split_col], axis=1)

# Display the DataFrame
print(df)
```

```python
# Set the first row as the column names
df.columns = df.iloc[0]

# Remove the first row
df = df[1:]

# Convert the 'Year' column to numeric
df['Year'] = pd.to_numeric(df['Year'])

# Display the DataFrame
print(df)
```

```python
# now insert the generated values into the dataset
df['Education'] = Education
df['Awareness'] = Awareness
df['Settlement'] = Settlement
df.head(10)
```

### Correlation matrix

```python
correlation_matrix = provinces[['Education', 'Awareness', 'Settlement', 'Free State',\
                                'Gauteng', 'Western Cape', 'Eastern Cape', 'KwaZulu-Natal','Limpopo',\
                                    'Mpumalanga','North West', 'Northern Cape']].corr()
print(correlation_matrix)
```

The values in the table range from -1 to 1.
A correlation of 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation.
In this matrix:
The diagonal represents the correlation of each variable with itself, which is always 1.
Off-diagonal elements show the correlation between different pairs of variables.
For example:
The correlation between Education and Awareness is 0.108995, indicating a weak positive correlation.
The correlation between Education and Settlement is 0.159070, also indicating a weak positive correlation.
The correlation between Awareness and Settlement is -0.126785, indicating a weak negative correlation.
The correlations between the temperature variables (Free State, Gauteng, etc.) and the other variables are also shown.

```python
#heatmap plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

Positive correlations suggest that as one variable increases, the other variable also tends to increase, while negative correlations suggest that as one variable increases, the other variable tends to decrease. A correlation close to 0 suggests no linear relationship between the variables

```python
# build Linear Regression model for regression analysis
import statsmodels.api as sm
#define independent and dependent variables
X= provinces[['Education', 'Awareness', 'Settlement', 'Free State',\
                                'Gauteng', 'Western Cape', 'Eastern Cape', 'KwaZulu-Natal','Limpopo',\
                                    'Mpumalanga','North West', 'Northern Cape']]
#add constant to the model
X = sm.add_constant(X)
#dependent variable
y = provinces['year']

#fit the model
model = sm.OLS(y, X).fit()
#summary of the model
print(model.summary())
```

Dependent Variable: 'year'
R-squared: The coefficient of determination, which measures the proportion of the variance in the dependent variable (year) that is predictable from the independent variables. In this case, it's 0.809, indicating that approximately 80.9% of the variance in the year variable is explained by the independent variables in the model.
Adj. R-squared: A version of the R-squared statistic that adjusts for the number of predictors in the model. It's 0.788 in this case.
F-statistic: A test statistic for the overall significance of the model. A higher F-statistic and a lower associated p-value indicate a better fit of the model. Here, the F-statistic is 38.14 with a very low p-value, indicating that the overall model is statistically significant.
Coefficients: These represent the estimated coefficients of the independent variables in the regression equation.
const: The intercept term of the regression equation.
Education, Awareness, Settlement: The coefficients for these variables represent the change in the dependent variable (year) for a one-unit change in each independent variable, holding all other variables constant. However, in this case, the p-values associated with these coefficients are relatively high, indicating that these variables are not statistically significant predictors of the year variable.
Provinces (Free State, Gauteng, Western Cape, etc.): The coefficients for these variables represent the change in the year variable associated with a one-unit change in each province, holding all other variables constant. For example, a one-unit increase in Gauteng is associated with an increase of 155.0903 in the year variable.
Standard Errors: Measure of the variability of the coefficient estimates. Lower standard errors indicate more precise estimates.
P>|t|: The p-values associated with each coefficient estimate. They indicate the probability of observing the estimated coefficient value (or more extreme) if the null hypothesis (that the true coefficient is zero) is true. Lower p-values (<0.05) indicate that the corresponding independent variable is statistically significant.
Confidence Intervals: Provide a range of plausible values for the true population coefficient with a certain level of confidence (usually 95%).
The notes at the bottom provide additional information and warnings about the regression analysis. In this case, it mentions the possibility of strong multicollinearity or other numerical problems due to a large condition number. Multicollinearity occurs when independent variables in a regression model are highly correlated with each other, which can cause issues with interpreting individual coefficients.

### Logistic Regression (GLM with Logit link function)

Logistic regression is suitable when the dependent variable is binary (in your case, public perception might be binary, with categories like positive/negative perception)

```python
# build logistic regression (GLM with logit link function)
import statsmodels.api as sm
#define independent and dependent variables
X= provinces[['Education', 'year', 'Settlement', 'Free State',\
                                'Gauteng', 'Western Cape', 'Eastern Cape', 'KwaZulu-Natal','Limpopo',\
                                    'Mpumalanga','North West', 'Northern Cape']]
#add constant to the model
X = sm.add_constant(X)

#dependent variable
y = provinces['Awareness']

#fit the model
model = sm.GLM(y, X, family=sm.families.Binomial()).fit()
#summary of the model
print(model.summary())
```

Dependent Variable: "Awareness"
Model Family: Binomial (indicating logistic regression)
Link Function: Logit (the log-odds are modeled as a linear combination of the predictors)
Method: IRLS (Iteratively Reweighted Least Squares)
Log-Likelihood: The log-likelihood of the model is -77.236.
Deviance: A measure of goodness of fit, which in this case is 154.47.
Pseudo R-squared: The McFadden pseudo R-squared value is 0.09642, indicating the proportion of variance explained by the model relative to the null model.
Coefficient Estimates: Each predictor variable has an associated coefficient estimate, along with its standard error, z-value, and p-value.
The coefficients represent the change in the log-odds of "Awareness" for a one-unit change in each predictor variable, holding other variables constant. The p-values associated with each coefficient test the null hypothesis that the coefficient is equal to zero (i.e., the predictor has no effect on the log-odds of "Awareness").
